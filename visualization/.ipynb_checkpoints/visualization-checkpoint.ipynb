```python
import xlrd
from datetime import date,datetime

file="D:\Downloads\\annual-co-emissions-by-region.xlsx"
file2="D:\Downloads\\hadcrut-surface-temperature-anomaly.xlsx"
file3="D:\Downloads\\temperature-anomaly.xlsx"
file4="D:\Downloads\\co2-concentration-long-term.xlsx"
file5="D:\Downloads\\average-real-gdp-per-capita-across-countries-and-regions.xlsx"

f=xlrd.open_workbook(filename=file)
ff=xlrd.open_workbook(filename=file2)
fff=xlrd.open_workbook(filename=file3)
ffff=xlrd.open_workbook(filename=file4)
fffff=xlrd.open_workbook(filename=file5)

sheet=f.sheet_by_index(0)
sheet2=ff.sheet_by_index(0)
sheet3=fff.sheet_by_index(0)
sheet4=ffff.sheet_by_index(0)
sheet5=fffff.sheet_by_index(0)
```

```python
temp={"Country name": [{"year": "Temperature"}]}
temp["temp"]=[{"temp": "year"}]
temp["temp"].append({"temp": "year"})

co2data={}
temdata={}
globaltem={}
globalco2={}
globalco2emi={}
gdppc={}

for i in range(1, sheet.nrows):
    row=sheet.row_values(i)
    if(row[0] not in co2data.keys()):
        co2data[row[0]]={}
    co2data[row[0]][row[2]]=row[3]
    
for i in range(1, sheet2.nrows):
    row=sheet2.row_values(i)
    if(row[0] not in temdata.keys()):
        temdata[row[0]]={}
    temdata[row[0]][row[2]]=row[3]

for i in range(1, sheet3.nrows):
    row=sheet3.row_values(i)
    globaltem[row[2]]=row[3]    

for i in range(1, sheet4.nrows):
    row=sheet4.row_values(i)
    globalco2[row[2]]=row[3] 

for i in range(1, sheet.nrows):
    row=sheet.row_values(i)
    if(row[2] not in globalco2emi.keys()):
        globalco2emi[row[2]]=0
    globalco2emi[row[2]]+=row[3]

for i in range(1, sheet5.nrows):
    row=sheet5.row_values(i)
    if(row[0] not in gdppc.keys()):
        gdppc[row[0]]={}
    gdppc[row[0]][row[2]]=row[3]


cleanglobaltem={}
for i in range(len(globalco2)):
    year=list(globalco2.keys())[i]
    if year+1 in globaltem.keys():
        cleanglobaltem[year]=globaltem[year+1] # use next year temperature but still name it last year
cleanglobalco2=globalco2.copy()
cleanglobalco2.popitem()

cleanglobalco2emi={}
for i in range(len(globaltem)):
    year=list(globaltem.keys())[i]
    if year-1 in globalco2emi.keys():
        cleanglobalco2emi[year]=globalco2emi[year-1]
#globalco2.popitem()
```

```python
co2index=sorted(list(cleanglobalco2.values()))
temindex=sorted(list(cleanglobaltem.values()))
size1=[]
size2=[]
for i in list(cleanglobalco2.values()):
    size1.append(co2index.index(i))

for i in list(cleanglobaltem.values()):
    size2.append(temindex.index(i))
    
df=pd.DataFrame({"year": list(cleanglobaltem.keys()), 
              "co2 concentration": list(cleanglobalco2.values()),
              "temperature anomaly": list(cleanglobaltem.values()),
                 "size1": size1,
                 "size2": size2
                })

####################################################
df.to_csv('tempcsv.csv')
```

```python
tree=[]
def treegenerator(region, cleancountryco2):
    tempco2=cleancountryco2.copy()
    treedata={}
    treedata["value"]=0
    treedata["name"]=region[0]
    treedata["children"]=[]
    for i in region:
        if i in list(tempco2.keys()):
            temp={}
            temp["name"]=i
            temp["value"]=sum(tempco2[i])
            treedata["value"]+=temp["value"]
            treedata["children"].append(temp)
    for i in treedata["children"]:
        i["value"]=i["value"]/treedata["value"]
    return treedata
for i in [africa, asia, na, sa, oceania]:
    tree.append(treegenerator(i,cleancountryco2))
print(tree)
############################################################
f=open('temptxt.txt','w')
f.write(str(tree))
f.close()

f=open("temptxt.txt",'r')
a=f.read()
tree=eval(a)
f.close()
############################################################
print(tree)

counter=0
for i in range(len(tree)):
    counter+=tree[i]["value"]

for i in range(len(tree)):
    tree[i]["value"]=tree[i]["value"]/counter

# change the ratio from region to global
for i in tree:
    for j in i["children"]:
        j["value"]=j["value"]*i["value"]
bcounter=round(counter/pow(10,9))
def labelformatter(params):
    if window.parseFloat(params.value*100).toFixed(2)>15:
        return params.name+"\n"+window.parseFloat(params.value*bcounter).toFixed(2)+r"billion tonnes co2 emission"
    else:
        return params.name+"\n"+window.parseFloat(params.value*100).toFixed(2)+r"%"+"co2 emission" 

treemap=TreeMap(width=800, height=800, title_text_size=24, is_animation=True, renderer="canvas")
treemap.add("Co2 emission in 10 years", tree, is_random=False, is_legend_show=True
           , line_opacity=0.5
           , is_label_show=True
           , label_pos="inside"
           , legend_orient="vertical"
           , label_formatter=labelformatter
           , is_more_utils=True
           , label_text_size =24)
#treemap.render()
```
